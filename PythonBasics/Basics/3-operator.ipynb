```python
a=10
b=20
#arithmetic operations
print("a + b =", a + b) ## addition
print("a - b =", a - b) ## subtraction
print("a * b =", a * b) ## multiplication
print("a / b =", a / b) ## division
print("a // b =", a // b) ## floor division
print("a % b =", a % b)  ## modulus
print("a ** b =", a ** b) ## exponentiation
#comparison operations
print("a == b:", a == b)  ## equal to
print("a != b:", a != b)  ## not equal to
print("a > b:", a > b)    ## greater than
print("a < b:", a < b)    ## less than
print("a >= b:", a >= b)  ## greater than or equal to
print("a <= b:", a <= b)  ## less than or equal to
#logical operations
print("a > 5 and b < 30:", a > 5 and b < 30)  ## logical AND
print("a < 5 or b < 30:", a < 5 or b < 30)    ## logical OR
print("not (a < 5):", not (a < 5))            ## logical NOT
#bitwise operations
print("a & b =", a & b)  ## bitwise AND
print("a | b =", a | b)  ## bitwise OR
print("a ^ b =", a ^ b)  ## bitwise XOR
print("a << 2 =", a << 2)  ## left shift
print("a >> 2 =", a >> 2)  ## right shift
#assignment operations
a += 5  # equivalent to a = a + 5
b -= 5  # equivalent to b = b - 5
```

```python
a=10
b=20
#identity operations
print("a is b:", a is b)  ## identity operator
print("a is not b:", a is not b)  ## identity operator
#membership operations
list1 = [10, 20, 30]
print("10 in list1:", 10 in list1)  ## membership operator
print("40 not in list1:", 40 not in list1)  ## membership operator

#ternary operation
result = "a is greater" if a > b else "b is greater"
print("Ternary operation result:", result)  ## ternary operation

#type casting
x = 5
y = 2.5
z = "10"
print("Integer to float:", float(x))  ## int to float
print("Float to integer:", int(y))  ## float to int
print("String to integer:", int(z))  ## str to int
print("Integer to string:", str(x))  ## int to str
```

```python
#string operations
str1 = "Hello"
str2 = "World"
print("Concatenation:", str1 + " " + str2)  ## string concatenation
print("Repetition:", str1 * 3)  ## string repetition
print("Length of str1:", len(str1))  ## length of string
print("Uppercase str1:", str1.upper())  ## convert to uppercase
print("Lowercase str2:", str2.lower())  ## convert to lowercase
print("str1[1:4]:", str1[1:4])  ## string slicing
print("str1.find('l'):", str1.find('l'))  ## find substring
print("str1.replace('l', 'x'):", str1.replace('l', 'x'))  ## replace substring
print("str1.split('l'):", str1.split('l'))  ## split string
print("str1.startswith('He'):", str1.startswith('He'))  ## check start of string
print("str2.endswith('ld'):", str2.endswith('ld'))  ## check end of string
print("str1.isalpha():", str1.isalpha())  ## check if all characters are alphabetic
print("str1.isdigit():", str1.isdigit())  ## check if all characters are digits
print("str1.isalnum():", str1.isalnum())  ## check if all characters are alphanumeric
print("str1.capitalize():", str1.capitalize())  ## capitalize first character
print("str1.title():", str1.title())  ## title case
print("str1.strip():", str1.strip())  ## remove leading and trailing whitespace
print("str1.center(10):", str1.center(10))  ## center align string
print("str1.ljust(10):", str1.ljust(10))  ## left align string
print("str1.rjust(10):", str1.rjust(10))  ## right align string
print("str1.count('l'):", str1.count('l'))  ## count occurrences of substring
print("str1.islower():", str1.islower())  ## check if all characters are lowercase
print("str1.isupper():", str1.isupper())  ## check if all characters are uppercase
print("str1.swapcase():", str1.swapcase())  ## swap case of characters
print("str1.partition('l'):", str1.partition('l'))  ## partition string
print("str1.zfill(10):", str1.zfill(10))  ## zero-fill string
print("str1.format('World'):", str1.format("World"))  ## format string
print("str1.join(['A', 'B','c']):", str1.join(['A', 'B','c']))  ## join strings
print("str1.translate(str.maketrans('l', 'x')):", str1.translate(str.maketrans('l', 'x')))  ## translate characters
print("str1.splitlines():", str1.splitlines())  ## split string into lines
```

```python
#dictionary operations
dict1 = {'a': 1, 'b': 2, 'c': 3}
print("Dictionary:", dict1)  ## print dictionary
print("Keys of dict1:", dict1.keys())  ## get keys
print("Values of dict1:", dict1.values())  ## get values
print("Items of dict1:", dict1.items())  ## get items
print("Get value for key 'b':", dict1.get('b'))  ## get value by key
print("Add new key-value pair 'd': 4:", dict1.update({'d': 4}))  ## add new key-value pair
print("Dictionary after adding 'd':", dict1)  ## print dictionary after adding
print("Remove key 'a':", dict1.pop('a'))  ## remove key-value pair
print("Dictionary after removing 'a':", dict1)  ## print dictionary after removing
print("Check if key 'b' exists:", 'b' in dict1)  ## check if key exists
print("Clear dictionary:", dict1.clear())  ## clear dictionary
print("Dictionary after clearing:", dict1)  ## print cleared dictionary
print("Copy of dictionary:", dict1.copy())  ## copy dictionary
print("Dictionary with default values:", dict.fromkeys(['x', 'y'], 0))  ## create dictionary with default values
print("Dictionary comprehension:", {k: v * 2 for k, v in dict1.items()})  ## dictionary comprehension
print("Merge two dictionaries:", {**dict1, **{'e': 5, 'f': 6}})  ## merge dictionaries
print("Dictionary as a list of tuples:", list(dict1.items()))  ## convert dictionary to list of tuples
print("Check if all keys are strings:", all(isinstance(k, str) for k in dict1.keys()))  ## check if all keys are strings
print("Check if all values are integers:", all(isinstance(v, int) for v in dict1.values()))  ## check if all values are integers
print("Dictionary with sorted keys:", {k: dict1[k] for k in sorted(dict1)})  ## dictionary with sorted keys
print("Dictionary with sorted values:", {k: v for k, v in sorted(dict1.items(), key=lambda item: item[1])})  ## dictionary with sorted values
print("Dictionary with keys as uppercase:", {k.upper(): v for k, v in dict1.items()})  ## dictionary with keys as uppercase
print("Dictionary with values as strings:", {k: str(v) for k, v in dict1.items()})  ## dictionary with values as strings
print("Dictionary with keys and values swapped:", {v: k for k, v in dict1.items()})  ## swap keys and values
print("Dictionary with keys as tuples:", {tuple(k): v for k, v in dict1.items()})  ## dictionary with keys as tuples
print("Dictionary with values as lists:", {k: [v] for k, v in dict1.items()})  ## dictionary with values as lists
print("Dictionary with keys as sorted tuples:", {tuple(sorted(k)): v for k, v in dict1.items()})  ## dictionary with keys as sorted tuples
print("Dictionary with values as sorted lists:", {k: sorted(v) for k, v in dict1.items()})  ## dictionary with values as sorted lists
print("Dictionary with keys as lowercase:", {k.lower(): v for k, v in dict1.items()})  ## dictionary with keys as lowercase
print("Dictionary with values as uppercase:", {k: v.upper() for k, v in dict1.items() if isinstance(v, str)})  ## dictionary with values as uppercase
print("Dictionary with keys as integers:", {int(k): v for k, v in dict1.items() if isinstance(k, str) and k.isdigit()})  ## dictionary with keys as integers
print("Dictionary with values as floats:", {k: float(v) for k, v in dict1.items() if isinstance(v, int)})  ## dictionary with values as floats
print("Dictionary with keys as sorted integers:", {k: v for k, v in sorted(dict1.items(), key=lambda item: int(item[0]) if item[0].isdigit() else float('inf'))})  ## dictionary with keys as sorted integers
print("Dictionary with values as sorted integers:", {k: sorted(v) for k, v in dict1.items() if isinstance(v, list)})  ## dictionary with values as sorted integers
print("Dictionary with keys as sorted strings:", {k: v for k, v in sorted(dict1.items(), key=lambda item: item[0])})  ## dictionary with keys as sorted strings
print("Dictionary with values as sorted strings:", {k: ''.join(sorted(v)) for k, v in dict1.items() if isinstance(v, str)})  ## dictionary with values as sorted strings
print("Dictionary with keys as sorted tuples:", {tuple(sorted(k)): v for k, v in dict1.items() if isinstance(k, (list, tuple))})  ## dictionary with keys as sorted tuples
print("Dictionary with values as sorted tuples:", {k: tuple(sorted(v)) for k, v in dict1.items() if isinstance(v, (list, tuple))})  ## dictionary with values as sorted tuples
print("Dictionary with keys as sorted lists:", {tuple(sorted(k)): v for k, v in dict1.items() if isinstance(k, list)})  ## dictionary with keys as sorted lists
print("Dictionary with values as sorted lists:", {k: sorted(v) for k, v in dict1.items() if isinstance(v, list)})  ## dictionary with values as sorted lists
```
